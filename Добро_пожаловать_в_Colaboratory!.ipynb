```
df.hist (column='Age', bins=40, color = 'green')
```

Построил график распределения возраста объектов, используя библиотеку pandas. По построенной гистограмме видно, что больше всего объектов с возрастом 23 года. Метод .hist в данном случае принимает 2 аргумента: column, то есть колонка, по данным которой строится гистограмма, и bins, в которой указывается количество столбцов гистограммы. По умолчанию цвет гистограммы синий, но мы можем изменить его с помощью аргумента color на, например, зеленый. Также можно убрать сетку с помощью аргумента grid и присвоить ему значение False, но это отрицательно повлияет на наглядность при работе с большим объемом данных.

```
plt.hist(df['Age'], edgecolor = 'black', bins = 40)
```

Построил гистограмму с помощью библиотеки matplotlib. Данная гистограмма кажется менее наглядной за счет отсутствия разметочной сетки. Так как данные для графика использовались те же, что и для предыдущего, с аналогичным числом столбцов (40), то он имеет аналогичный вид. Единственное отличие - черное обрамление, заложенное в переменной edgecolor.

```
go.Figure(data=[go.Histogram(x=df['Age'])]).update_layout(legend_orientation="h",
                  hovermode="x",
                  height=600,
                  width=800)

# https://habr.com/ru/articles/502958/
```

Построил гистограмму с помощью метода graph_objs библиотеки plotly, импортированного как go. Данные те же - вывод тот же. Переменная hovermode отвечает за отображение значений в конкретной точке, выбранной курсором, на оси. В данном случае выбрана ось абсцисс. Переменные height и width отвечают за высоту и ширину выводимой гистограммы.

Построил объект boxplot, используя библиотеку seaborn. x - данные, отображаемые по оси абсцисс, y - данные, отображаемые по оси ординат, orient - ориентация боксплота (в данном случае горизонтальная)
https://seaborn.pydata.org/generated/seaborn.boxplot.html
Исходя из построенного боксплота, основной возраст пассажиров 1 класса - от 26 до 48 лет, пассажиров 2 класса - от 22 до 35 лет с выбросами (нетипичными единичными значениями) от 0 до 4 лет и от 55 до 72 лет, и пассажиров 3 класса - от 18 до 32 лет с выбросами от 53 до 75 лет.
