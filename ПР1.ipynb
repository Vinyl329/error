Построил объект boxplot, используя библиотеку seaborn. x - данные, отображаемые по оси абсцисс, y - данные, отображаемые по оси ординат, orient - ориентация боксплота (в данном случае горизонтальная)
https://seaborn.pydata.org/generated/seaborn.boxplot.html
Исходя из построенного боксплота, основной возраст пассажиров 1 класса - от 26 до 48 лет, пассажиров 2 класса - от 22 до 35 лет с выбросами (нетипичными единичными значениями) от 0 до 4 лет и от 55 до 72 лет, и пассажиров 3 класса - от 18 до 32 лет с выбросами от 53 до 75 лет.

```
numeric_col = ['Survived', 'Age']
corr_matrix = df.loc[:,numeric_col].corr()
sns.heatmap(df.corr(), annot = True, fmt='.1g', cmap= 'coolwarm')
```

Построил корреляционную матрицу с помощью библиотеки seaborn. Переменная annot отвечает за аннотацию каждой ячейки, по умолчанию она равна False. Переменная fmt отвечает за формат аннотации при выводе. Например, если бы она была равна '.4%', та данные в каждой ячейке выводились бы в виде процентов с 4 знаками после запятой. Переменная cmap отвечает за цветовую гамму при выводе.

Значение 1 означает положительную линейную зависимость, то есть, если одна переменная увеличивается, то и другая переменная также увеличивается. Значение -1 означает отрицательную линейную зависимость, то есть, если одна переменная увеличивается, то другая переменная уменьшается. Значение 0 означает отсутствие линейной зависимости между переменными.
